[Problem 5] Terminology research
1. What is kurtosis?
Kurtosis is a statistical measure used to describe a characteristic of a dataset. When normally distributed data is plotted on a graph, it generally takes the form of an upsidedown bell. This is called the bell curve. The plotted data that are furthest from the mean of the data usually form the tails on each side of the curve. Kurtosis indicates how much data resides in the tails.

2. What is skew?
Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean

[Question 7] Check the correlation coefficient 
Create a heat map of the correlation coefficient matrix and find relationships with high correlation coefficients.

```python
import matplotlib.pyplot as plt
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

Select 10 features that are highly correlated with the target variable, and create a correlation coefficient matrix heat map for just these.

```python
k = 10 
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
