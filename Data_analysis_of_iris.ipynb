[Problem 4] Investigation of the iris dataset itself At the beginning of the page, I wrote "On a pasture on the Gaspé Peninsula, Canada", but let's explore the dataset itself a little more. In looking at a dataset, it's also important to know about the dataset itself.

The Iris flower data set or Fisher's Iris data set is a multivariate data set used and made famous by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis. This is perhaps the best known database to be found in the pattern recognition literature. Fisher's paper is a classic in the field and is referenced frequently to this day. (See Duda & Hart, for example.) The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.
